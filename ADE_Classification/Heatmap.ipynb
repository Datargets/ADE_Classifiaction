```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

# Define three different confusion matrices
confusion_matrices = [
    np.array([
        [1644, 27],
        [125, 294]
    ]),
    np.array([
        [1618, 53],
        [100, 319]
    ]),
    np.array([
        [1545, 126],
        [56, 363]
    ]),

]

# Define the labels for the confusion matrices
labels = ['ADE', 'Non-ADE']

# Create a custom colormap from white to light gray
cmap = mcolors.LinearSegmentedColormap.from_list("light_gray", ["white", "lightgray"])

# Create the figure with 3 subplots
fig, axes = plt.subplots(1, 3, figsize=(11, 3))

# Titles for the subplots
titles = ['Bert Fine-Tune with LoRA', 'DistilBert Fine-Tune with LoRA', 'RoBerta Fine-Tune with LoRA']

# Loop through each subplot and corresponding confusion matrix
for i, (conf_matrix, ax) in enumerate(zip(confusion_matrices, axes)):
    sns.heatmap(conf_matrix, annot=True, fmt="d", cmap=cmap,
                xticklabels=labels, yticklabels=labels, ax=ax)
    ax.set_title(titles[i])
    ax.set_xlabel('Predicted label')
    ax.set_ylabel('True label')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data from the table
models = ['BERT-base-cased', 'RoBERTa-base', 'Distil-BERT-uncased']
classes = ['Non-ADE:0', 'ADE:1']
precision = np.array([[0.93, 0.92], [0.97, 0.74], [0.94, 0.86]])
recall = np.array([[0.98, 0.70], [0.92, 0.87], [0.97, 0.76]])
f1_score = np.array([[0.96, 0.79], [0.94, 0.80], [0.95, 0.81]])

# Create a figure and axis
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Bar plot for Precision, Recall, and F1 Score
metrics = [precision, recall, f1_score]
metric_names = ['Precision', 'Recall', 'F1 Score']
colors = ['lightblue', 'blue', 'gray']

for i, metric in enumerate(metrics):
    ax = axes[i]
    width = 0.35
    bar_width = 0.25
    x = np.arange(len(models))

    ax.bar(x - bar_width/2, metric[:, 0], bar_width, label='Non-ADE:0', color='lightblue')
    ax.bar(x + bar_width/2, metric[:, 1], bar_width, label='ADE:1', color='blue')

    ax.set_xticks(x)
    ax.set_xticklabels(models)
    ax.set_ylabel(metric_names[i])
    ax.set_title(f'{metric_names[i]} for Different Models')
ax.legend(loc='lower left', bbox_to_anchor=(1, 0.5))
()

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data from the table
models = ['Bert-base-cased', 'Roberta-base', 'Distil-bert-uncased']

accuracy = [0.927, 0.912, 0.926]
matthew_corrcoef = [0.761, 0.747, 0.763]
avg_precision = [0.702, 0.66, 0.70]

# Plotting
x = np.arange(len(models))
width = 0.1  # Bar width

fig, ax = plt.subplots(figsize=(10, 6))

# Define colors in range from blue to gray
colors = ['#4F81BD', '#A9A9A9', '#ADD8E6']

# Plotting the bars for each metric
ax.bar(x, accuracy, width, label='Accuracy', color=colors[2])
ax.bar(x + width, matthew_corrcoef, width, label='Matthew Corrcoef', color=colors[0])
ax.bar(x + 2*width, avg_precision, width, label='Avg Precision', color=colors[1])

# Formatting the plot
ax.set_xlabel('Models')
ax.set_ylabel('Scores')
ax.set_title('Model Performance Metrics')
ax.set_xticks(x)
ax.set_xticklabels(models)
ax.legend(loc='best', bbox_to_anchor=(1, 0.5))

plt.tight_layout()
plt.show()
```
