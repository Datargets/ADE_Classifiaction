```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

# Define three different confusion matrices
confusion_matrices = [
    np.array([
        [1629, 42],
        [116, 303]
    ]),
    np.array([
        [1586, 85],
        [86, 333],

    ]),
    np.array([
        [1589, 82],
        [68, 351]
    ]),

]

labels = ['ADE', 'Non_ADE0']

cmap = mcolors.LinearSegmentedColormap.from_list("light_gray", ["white", "lightgray"])

fig, axes = plt.subplots(1, 3, figsize=(12, 3))

titles = ['Bert Confusion Matrix', 'DistilBert Confusion Matrix', 'Roberta Confusion Matrix']

for i, (conf_matrix, ax) in enumerate(zip(confusion_matrices, axes)):
    sns.heatmap(conf_matrix, annot=True, fmt="d", cmap=cmap,
                xticklabels=labels, yticklabels=labels, ax=ax)
    ax.set_title(titles[i])
    ax.set_xlabel('Predicted label')
    ax.set_ylabel('True label')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Updated Data from the table
models = ['BERT-base-cased', 'RoBERTa-base', 'Distil-BERT-uncased']
classes = ['Non-ADE:0', 'ADE:1']
precision = np.array([[0.93, 0.88], [0.96, 0.81], [0.95, 0.80]])
recall = np.array([[0.97, 0.72], [0.95, 0.84], [0.95, 0.79]])
f1_score = np.array([[0.95, 0.79], [0.95, 0.82], [0.95, 0.80]])

# Create a figure and axis
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Bar plot for Precision, Recall, and F1 Score
metrics = [precision, recall, f1_score]
metric_names = ['Precision', 'Recall', 'F1 Score']
colors = ['lightblue', 'blue']

for i, metric in enumerate(metrics):
    ax = axes[i]
    width = 0.35
    bar_width = 0.25
    x = np.arange(len(models))

    ax.bar(x - bar_width/2, metric[:, 0], bar_width, label='Non-ADE:0', color='lightblue')
    ax.bar(x + bar_width/2, metric[:, 1], bar_width, label='ADE:1', color='blue')

    ax.set_xticks(x)
    ax.set_xticklabels(models)
    ax.set_ylabel(metric_names[i])
    ax.set_title(f'{metric_names[i]} for Different Models')

# Set legend outside the last plot
axes[-1].legend(loc='lower left', bbox_to_anchor=(1, 0.5))

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

models = ['Bert-base-cased', 'Roberta-base', 'Distil-bert-uncased']

accuracy = [0.924, 0.928, 0.918]
matthew_corrcoef = [0.752, 0.779, 0.744]
avg_precision = [0.69, 0.711, 0.674]

x = np.arange(len(models))
width = 0.15

fig, ax = plt.subplots(figsize=(12, 6))

colors = ['#4F81BD', '#A9A9A9', '#ADD8E6']

ax.bar(x, accuracy, width, label='Accuracy', color=colors[2])
ax.bar(x + width, matthew_corrcoef, width, label='Matthew Corrcoef', color=colors[0])
ax.bar(x + 2*width, avg_precision, width, label='Avg Precision', color=colors[1])

ax.set_xlabel('Models')
ax.set_ylabel('Scores')
ax.set_title('Model Performance Metrics')
ax.set_xticks(x)
ax.set_xticklabels(models)
ax.legend(loc='best', bbox_to_anchor=(1, 0.5))

plt.tight_layout()
plt.show()
```
